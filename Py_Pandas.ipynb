```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'A','B'],
    'Value': [10, 15, 20, 25, 30, 35, 40, 45,50]
}
df = pd.DataFrame(data)
df
```

```python
# Example 1: Group by 'Category' and calculate the mean of 'Value'
m=df.groupby('Category').Value.mean().reset_index()# becomes regular df
m.columns = ['Category', 'Mean_Values']
#mm = m.to_frame(name='Mean_Value_for_Category')# series to df title
mm
```

```python
#Example 2: Group by 'Category' and calculate the sum of 'Value'
grouped_sum = df.groupby('Category').Value.sum().reset_index()
grouped_sum
```

```python
# Example 3: Group by multiple columns and calculate the mean
multi_grouped_mean = df.groupby(['Category', df.index % 2])['Value'].mean()
multi_grouped_mean
```

```python
# Example 4: Group by 'Category' and describe(general statistics) for each category
grouped_describe = df.groupby('Category')['Value'].describe()
grouped_describe
```

```python
#Example 5: Apply multiple aggregation functions simultaneously
agg_functions = {'Value': ['mean', 'sum', 'min', 'max']}
grouped_multiple_agg = df.groupby('Category').agg(agg_functions)
grouped_multiple_agg
```

```python
# Example 6: Group by 'Category' and count the number of occurrences
grouped_count = df.groupby('Category')['Value'].count()#same for Category.count()
grouped_count
```

```python
#Example 7: Group by 'Category' and get the first element of each group
grouped_first = df.groupby('Category').first()
grouped_first
```

```python
#Example 8: Group by 'Category' and get the nth element of each group
grouped_nth_element = df.groupby('Category').nth(1)#2nd element
grouped_nth_element
```

```python
# Example 9: Group by 'Category' and apply a custom aggregation function
custom_agg_function = lambda x: np.sum(x) / np.mean(x)
custom_agg_result = df.groupby('Category')['Value'].agg(custom_agg_function)
custom_agg_result
```

```python
#Example 10: Group by 'Category' and apply a custom function to each group
def custom_function(group):
    return group['Value'] - group['Value'].mean()

custom_result = df.groupby('Category').apply(custom_function)
custom_result
```

```python
# Example 11: Group by 'Category' and filter groups based on a condition
grouped_filter = df.groupby('Category').filter(lambda x: x['Value'].mean() > 25)
grouped_filter
```

```python
# Example 12: Group by 'Category' and transform the values within each group
grouped_transform = df.groupby('Category')['Value'].transform(lambda x: x - x.mean())
grouped_transform
```

```python
#Example 13: Group by 'Category' and apply a rolling window operation
rolling_avg = df.groupby('Category')['Value'].rolling(window=3).mean()#taking 3 cons values to calculate moving avg
rolling_avg
```

```python
# Example 14: Group by 'Category' and apply a cumulative sum within each group
grouped_cumsum = df.groupby('Category')['Value'].cumsum()#cumsums when it finds value of it type.
grouped_cumsum
```

```python
#Example 15: Group by 'Category' and get the Category with the maximum 'Value'
grouped_max_group = df.loc[df.groupby('Category')['Value'].idxmax()]#locates max index
grouped_max_group
```

```python
# Example 16: Group by 'Category' and get the top N values within each group
top_n_values = df.groupby('Category')['Value'].nlargest(2)
top_n_values
```

```python
# Example 17: Group by 'Category' and get the rank of each value within each group
grouped_rank = df.groupby('Category')['Value'].rank()
grouped_rank
```

```python
# Example 18: Group by 'Category' and apply a shift operation within each group
grouped_shift = df.groupby('Category')['Value'].shift()#shifts one step down, NAN for 1st value for each category
grouped_shift
```

```python
# Example 19: Group by 'Category' and apply a pivot table operation
pivot_table_result = df.pivot_table(index='Category', columns=df.index % 2, values='Value', aggfunc='mean')
pivot_table_result# mean value for odd and even index in each category
```

```python
# Example 20: Group by 'Category' and apply a custom aggregation function using agg
agg_custom_result = df.groupby('Category').agg({'Value': lambda x: np.sum(x) / np.mean(x)})
agg_custom_result
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, 30, 22, 35, 22],
    'Salary': [50000, 60000, 45000, 70000, 55000],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR']
}
df = pd.DataFrame(data)
df
```

```python
# Example 1: Sort by a single column (ascending)
sorted_by_age_asc = df.sort_values(by='Age',ascending=True)
sorted_by_age_asc
```

```python
# Example 2: Sort by a single column (descending)
sorted_by_age_desc = df.sort_values(by='Age', ascending=False)
sorted_by_age_desc
```

```python
# Example 3: Sort by multiple columns (ascending)
sorted_by_age_salary_asc = df.sort_values(by=['Age', 'Salary'])#within each age, it sorts by the 'Salary'
sorted_by_age_salary_asc
```

```python
# Example 4: Sort by multiple columns (descending)
sorted_by_age_salary_desc = df.sort_values(by=['Age', 'Salary'], ascending=[True, False])# age asc & salary des
sorted_by_age_salary_desc
```

```python
import matplotlib.pyplot as plt

# Data
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# Creating the bar plot
fig, ax = plt.subplots()

bars = ax.bar(fruits, counts, color=bar_colors)

# Setting labels for each bar
for bar, label in zip(bars, bar_colors):
    bar.set_label(label)

ax.set_ylabel('Fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2

# Two signals with a coherent part at 10 Hz and a random part
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel('Coherence')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

# some random data
x = np.random.randn(1000)
y = np.random.randn(1000)


def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
```

```python
# Start with a square Figure.
fig = plt.figure(figsize=(6, 6))
# Add a gridspec with two rows and two columns and a ratio of 1 to 4 between
# the size of the marginal axes and the main axes in both directions.
# Also adjust the subplot parameters for a square plot.
gs = fig.add_gridspec(2, 2,  width_ratios=(4, 1), height_ratios=(1, 4),
                      left=0.1, right=0.9, bottom=0.1, top=0.9,
                      wspace=0.05, hspace=0.05)
# Create the Axes.
ax = fig.add_subplot(gs[1, 0])
ax_histx = fig.add_subplot(gs[0, 0], sharex=ax)
ax_histy = fig.add_subplot(gs[1, 1], sharey=ax)
# Draw the scatter plot and marginals.
scatter_hist(x, y, ax, ax_histx, ax_histy)
```
