```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation
data = {
    'year': [2017, 2018, 2019, 2020, 2021],
    'value': [10, 15, 20, 18, 25]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))  # Set the figure size
df.plot(x='year', y='value', kind='line')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trend of Value Over Years')
plt.show()
```

```python
# Sample DataFrame creation
data = {
    'category': ['A', 'B', 'C', 'D'],
    'count': [30, 40, 25, 35]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))  # Set the figure size
df.plot(x='category', y='count', kind='bar', rot=45)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of Categories')
plt.show()
```
