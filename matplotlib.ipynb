```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation
data = {
    'year': [2017, 2018, 2019, 2020, 2021],
    'value': [10, 15, 20, 18, 25]
}
df = pd.DataFrame(data)

plt.figure(figsize=(6, 6))  # Set the figure size
df.plot(x='year', y='value', kind='line')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trend of Value Over Years')
plt.show()
```

```python
# Sample DataFrame creation
data = {
    'category': ['A', 'B', 'C', 'D'],
    'count': [30, 40, 25, 35]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))  # Set the figure size
df.plot(x='category', y='count', kind='bar', rot=90)#plt.bar() for ind data
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of Categories')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x_values = ['Category A', 'Category B', 'Category C', 'Category D']
y_values = [10, 20, 15, 25]

plt.figure(figsize=(8, 5))
plt.bar(x_values, y_values)
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Rotated Labels')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation
data = {
    'category': ['A', 'B', 'C', 'D'],
    'count': [25, 35, 45, 35]
}
df = pd.DataFrame(data)

# Create bins for count values
bins = [0, 30, 40, 50]  # Define bins(3-intervals)
labels = ['Low', 'Medium', 'High']  # Define corresponding labels

# Bin(=interval) the 'count' values
df['labeled_category'] = pd.cut(df['count'], bins=bins, labels=labels)#new col
print(df)
plt.figure(figsize=(10, 6))  # Set the figure size
df_category_counts = df['labeled_category'].value_counts().sort_index()
print(df_category_counts)
df_category_counts.plot(kind='bar', rot=0, color='skyblue')
plt.xlabel('Count Category')
plt.ylabel('Count')
plt.title('Count Categories')
plt.show()
```
